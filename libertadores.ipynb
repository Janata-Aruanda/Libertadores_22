```python
import pandas as pd 
#Importando a tabela
libertadores = pd.read_csv('Libertadores_Matches.csv')
#Visualizando a Tabela (Antes de renomear )
display(libertadores)
```

```python
#Renomeando nome de colunas 
libertadores.rename(columns={'datetime' : 'Data/Hora',
                       'home_team' : 'Time_Casa',
                        'away_team' : 'Time_Fora',  
                       'home_goal' : 'Gol_Casa',
                       'away_goal' : "Gol_Fora",
                        'season': 'Ano' , 'stage' : 'Fase_Grupo'}, inplace=True)
```

```python
#Visualizando a tabela (depois de renomear )
libertadores_22 = pd.DataFrame(libertadores)

display(libertadores_22)
```

```python
#Localizando o ano de 2022 e Salvando com "libertadores_2022 "
libertadores_2022 = libertadores_22[libertadores_22['Ano']==2022.0]  
```

```python
#Exibindo a litração do ano de 2022
display(libertadores_2022)
```

```python
#Resetar o index da tabela 
libertadores_2022.reset_index(inplace=True)
```

```python
#excluindo o Index que foi adicionado
libertadores_2022.drop('index' , axis=1, inplace=True )
```

```python
display(libertadores_2022)
```

```python
#Subistituindo os dados da coluna 
libertadores_2022['Fase_Grupo'].replace('group stage', 'fase de grupos', inplace=True)
libertadores_2022['Fase_Grupo'].replace('round of 16', 'oitavas de grupos', inplace=True)
libertadores_2022['Fase_Grupo'].replace('quarterfinals', 'quartas de grupos', inplace=True)
libertadores_2022['Fase_Grupo'].replace('semifinals', 'semifinal', inplace=True)
```

```python
#Subistituindo os dados da coluna 
libertadores_2022['Time_Casa'].replace('Athletico', 'Athletico-PR', inplace=True)
libertadores_2022['Time_Fora'].replace('Athletico', 'Athletico-PR', inplace=True)
```

```python
df = pd.DataFrame (libertadores_2022)
```
